### 2nd situation

```python
phase.set_index('Year')
```

```python
a = pd.pivot_table(phase, columns=['Year'])
a
```

```python
mean_2000 = a[2000].mean()
mean_2019 = a[2019].mean()

var_2000 = a[2000].var()
var_2019 = a[2019].var()

n_2000 = a[2000].sum()
n_2019 = a[2019].sum()
```

```python
dif_means = mean_2019 - mean_2000
se = np.sqrt(var_2000/n_2000 + var_2019/n_2019)
t_statstic = dif_means / se
t_statstic
```

```python
deg_free = np.min([n_2000, n_2019]) - 1

p_val = stats.t.cdf(t_statstic, deg_free)
```

```python
p_val = p_val*2
p_val
```

```python
alpha = 0.05
```

```python
p_val < alpha
```

```python
#We can conclude, that there is significance difference between both years. 
```
